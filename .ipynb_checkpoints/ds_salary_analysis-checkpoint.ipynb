```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("indeed_ds_jobs.csv")
df = df[df.Salary != "None"]
df
```

```python
salary = df['Salary'].apply(lambda x: x.split('a')[0])
remove_sign = salary.apply(lambda x: x.replace('$', ''))
salary_formatted = remove_sign.apply(lambda x: x.replace(',', '').replace(' ', ''))
df['Minimum Salary'] = salary_formatted.apply(lambda x: float((x.split('-'))[0]))
df['Maximum Salary'] = salary_formatted.apply(lambda x: float((x.split('-'))[-1]))
df['Average Salary'] = (df['Minimum Salary'] + df['Maximum Salary']) / 2
df
```

```python
def position(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data science' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'software engineer' in title.lower():
        return 'software engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'analytics' in title.lower():
        return 'analyst'
    elif 'data specialist' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'machine learning expert'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'research' in title.lower():
        return 'researcher'
    elif 'researcher' in title.lower():
        return 'researcher'
    else:
        return 'na'
```

```python
df['Position'] = df['Title'].apply(position)
df['Position'].value_counts()
```
